```python
directory=r"E:\Autogluon"
#time_limits = 3600*10 # for quick demonstration only, you should set this to longest time you are willing to wait (in seconds)

excluded_model_types = ['KNN', 'XT', 'RF',"LR"]
presets="medium_quality_faster_train" # roduces less accurate models but facilitates faster prototyping

#presets="best_quality"
hyperparameters = "light" 


predictor = TabularPredictor(
    label="unidades_vendidas", 
    time_limits=60*15,
    eval_metric="root_mean_squared_error",
    problem_type="regression"
)

predictor.fit(
    train_data,
    hyperparameters=hyperparameters,
    presets=presets,
    excluded_model_types=excluded_model_types
)
```
